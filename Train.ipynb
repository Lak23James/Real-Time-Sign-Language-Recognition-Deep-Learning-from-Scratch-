Desgining The algorithm 
All the Steps required for Gradient descent Algorithm

```python
#inliatlizing the varaibles
def init_params():
    W1 = np.random.rand(64, 784) - 0.5
    b1 = np.random.rand(64, 1) - 0.5
    W2 = np.random.rand(25, 64) - 0.5
    b2 = np.random.rand(25, 1) - 0.5
    return W1, b1, W2, b2

def ReLU(Z):
    return np.maximum(Z, 0)
# function to convert the Numbers Into bayesian stats
def softmax(Z):
    A = np.exp(Z) / sum(np.exp(Z))
    return A
# calculating the values using Matrix Multiplication
def forward_prop(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2
# Non Linearity Function
def ReLU_deriv(Z):
    return Z > 0
#Probability Matrix
def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, 25))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y
# finding the error betweem the true value and predicted value
def backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y):
    one_hot_Y = one_hot(Y)
    dZ2 = A2 - one_hot_Y
    dW2 = 1 / m * dZ2.dot(A1.T)
    db2 = 1 / m * np.sum(dZ2,axis=1, keepdims=True)
    dZ1 = W2.T.dot(dZ2) * ReLU_deriv(Z1)
    dW1 = 1 / m * dZ1.dot(X.T)
    db1 = 1 / m * np.sum(dZ1 ,axis=1, keepdims=True)
    return dW1, db1, dW2, db2
# Updating the Values of the varaiables
def update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    W1 = W1 - alpha * dW1
    b1 = b1 - alpha * db1    
    W2 = W2 - alpha * dW2  
    b2 = b2 - alpha * db2    
    return W1, b1, W2, b2
```

```python
# returning the predictions
def get_predictions(A2):
    return np.argmax(A2, 0)
# getting the accuracy 
def get_accuracy(predictions, Y):
    print(predictions, Y)
    return np.sum(predictions == Y) / Y.size
# making the steps in the Chornological order
def gradient_descent(X, Y, alpha, iterations):
    W1, b1, W2, b2 = init_params()
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
        dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y)
        W1, b1, W2, b2 = update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        if i % 10 == 0:
            print("Iteration: ", i)
            predictions = get_predictions(A2)
            print(get_accuracy(predictions, Y))
    return W1, b1, W2, b2

W1, b1, W2, b2 = gradient_descent(X_train, Y_train, 0.10, 5000)
```

```python
def make_predictions(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions

def test_prediction(index, W1, b1, W2, b2):
    current_image = X_test[:, index, None]
    prediction = make_predictions(X_test[:, index, None], W1, b1, W2, b2)
    label = Y_test[index]
    
    # The Alphabet Map (0=A, 1=B...)
    # Note: This dataset skips J (9) because it involves motion.
    # If your dataset maps 0->A directly, this works. 
    letters = "ABCDEFGHIKLMNOPQRSTUVWXY" # 24 letters (skipping J and Z)
    
    print("Prediction: ", prediction[0])
    print("True Label: ", label)
    
    # Try to convert to letter if possible
    if label < len(letters):
        print(f"Model says: {letters[prediction[0]]}")
        print(f"Actual is:  {letters[label]}")
        
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()

# Run this line to see a random example!
# Change the number '5' to any number between 0 and the size of your test set
test_prediction(5, W1, b1, W2, b2)
```
